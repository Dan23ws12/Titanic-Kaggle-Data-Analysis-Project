```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mat
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_excel(r"C:\Users\danie\Documents\data projects\Titanic data analysis\Train Data\train2.xlsx")
titanic = titanic.dropna()
```

```python
class_bins, surv_bins = [0, 1, 2, 3], [-1, 0, 1]
class_names, surv_names = ["First Class", "Second Class", "Third Class"], ["Didn't Survive", "Survived"]
titanic["Pclass"] = pd.cut(titanic["Pclass"], bins = class_bins, labels = class_names) #make Pclass categorical variable
titanic["Survived"] = pd.cut(titanic["Survived"], bins = surv_bins, labels = surv_names) #make Survived categorical variable
```

```python
titanic.head(10)
```

```python
#histogram of survival per passenger class
sns.countplot(data=titanic, x="Survived", hue="Pclass")
```

```python
#pie chart showing percent of people in each passenger class
ppl_pclass = titanic.groupby("Pclass").count()
plt.pie(ppl_pclass["Survived"], labels = titanic["Pclass"].unique())
plt.show()
```
